```python
from huggingface_hub import HfApi, hf_hub_download
from pathlib import Path
```

```python
REPO_ID = "tensorblock/BgGPT-Gemma-2-2.6B-IT-v1.0-GGUF"
LOCAL_DIR = Path(r"C:\models\bggpt-2.6b")
LOCAL_DIR.mkdir(parents=True, exist_ok=True)
```

```python
api = HfApi()
files = api.list_repo_files(repo_id=REPO_ID, repo_type="model")
candidates = [f for f in files if f.lower().endswith(".gguf") and "q4_k_m" in f.lower()]

if not candidates:
    # fallback: ако в repo-то е именувано малко по-различно
    candidates = [f for f in files if f.lower().endswith(".gguf") and "q4" in f.lower()]

if not candidates:
    raise RuntimeError("Не намерих gguf файл (Q4*) в repo-то. Провери repo files на HuggingFace.")
```

```python
filename = sorted(candidates)[0]
print("Downloading:", filename)
```

```python
local_path = hf_hub_download(
    repo_id=REPO_ID,
    filename=filename,
    local_dir=str(LOCAL_DIR),
    local_dir_use_symlinks=False,
)
```
